#Load the webpage & Store the scraped data in a python dictionary

```python
import requests
from bs4 import BeautifulSoup as bs

r = requests.get("https://en.wikipedia.org/wiki/Toy_Story_4")

#convert to beautiful soup object
soup = bs(r.content)
contents = soup.prettify()
contents
```

```python
info_box = soup.find(class_='infobox vevent')
#print(info_box.prettify())

#find rows in the info box table
info_rows=info_box.findAll("tr")
for row in info_rows:
    print(row.prettify())
```

#Save the above rows into a python dict

```python

#Check whether row has mutiple entries/single entry
def get_content_value(row_data):
    if row_data.find("li"):
        return[li.get_text() for li in row_data.find_all("li")]
    else:
        return row_data.get_text()
    
movie_info = {}

for index, row in enumerate(info_rows):
    if index == 0:
        movie_info['title']=row.find('th').get_text()
    elif index == 1:
        continue
    else:
        content_key = row.find('th').get_text()
        #content_value = row.find('td').get_text()
        content_value = get_content_value(row.find('td'))
        movie_info[content_key] =content_value
    
movie_info
```
